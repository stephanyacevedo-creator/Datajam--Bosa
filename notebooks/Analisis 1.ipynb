```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Carga y Filtrado de Datos para la Localidad de Bosa

```python
# Ruta del archivo CSV local especificado por el usuario
local_csv_path = '/content/drive/MyDrive/residuos /residuossolidos_-eficiencia1 (2).csv'

df_local_csv = pd.DataFrame() # Inicializar DataFrame vacío

# Lista de codificaciones a probar
encodings_to_try = ['utf-8', 'latin1', 'cp1252', 'utf-8-sig']

for encoding in encodings_to_try:
    try:
        print(f"Intentando cargar el archivo con codificación: {encoding}")
        # Intentar cargar el CSV con el separador ';' y engine='python' para manejar posibles problemas
        df_local_csv = pd.read_csv(local_csv_path, sep=';', engine='python', encoding=encoding)

        print(f"Datos cargados exitosamente desde: {local_csv_path} con codificación {encoding}")

        print("\nPrimeras 5 filas del DataFrame local:")
        display(df_local_csv.head())

        print("\nInformación general del DataFrame local:")
        df_local_csv.info()
        break # Si la carga es exitosa, salir del bucle

    except FileNotFoundError:
        print(f"Error: El archivo no se encontró en la ruta: {local_csv_path}")
        break # No tiene sentido probar otras codificaciones si el archivo no existe
    except UnicodeDecodeError as e:
        print(f"Fallo la decodificación con {encoding}: {e}")
        continue # Intentar la siguiente codificación
    except Exception as e:
        print(f"Ocurrió un error al cargar o analizar el archivo CSV local con {encoding}: {e}")
        break # Otro tipo de error, salir del bucle

if df_local_csv.empty:
    print("\nNo fue posible cargar el archivo CSV con ninguna de las codificaciones probadas.")
```

```python
locality_column = '3. ¿En qué Localidad vive?'

df_bosa_local = pd.DataFrame() # Initialize df_bosa_local to prevent NameError

# Ensure df_local_csv is not empty before attempting to filter
if not df_local_csv.empty and locality_column in df_local_csv.columns:
    # Filtrar datos para la localidad 'Bosa'.
    # Using .str.contains('Bosa', case=False, na=False) is robust.
    df_bosa_local = df_local_csv[df_local_csv[locality_column].astype(str).str.contains('Bosa', case=False, na=False)]

    if not df_bosa_local.empty:
        print("Primeras 5 filas de los datos de la localidad Bosa (archivo local):")
        display(df_bosa_local.head())
        print(f"\nNúmero total de registros para la localidad Bosa (archivo local): {len(df_bosa_local)}")

        response_column = '4. Los recibos de servicios públicos que recibe en su hogar, corresponden a: '
        if response_column in df_bosa_local.columns:
            print(f"\nDistribución de '{response_column}' en Bosa (archivo local):")
            response_counts = df_bosa_local[response_column].astype(str).value_counts()
            display(response_counts)

            plt.figure(figsize=(10, 6))
            sns.barplot(y=response_counts.index, x=response_counts.values, palette='crest')
            plt.title(f"Distribución de '{response_column}' en Bosa (Archivo Local)")
            plt.xlabel('Conteo')
            plt.ylabel(response_column)
            plt.grid(axis='x', linestyle='--', alpha=0.7)
            plt.tight_layout()
            plt.show()
        else:
            print(f"La columna '{response_column}' no se encontró en los datos de Bosa para análisis detallado. Columnas disponibles:")
            print(df_bosa_local.columns.tolist())
    else:
        print(f"No se encontraron registros para la localidad 'Bosa' en la columna '{locality_column}' del archivo local.")
elif df_local_csv.empty:
    print("No hay datos cargados del archivo local para filtrar la localidad Bosa.")
else:
    print(f"La columna '{locality_column}' no se encontró en el DataFrame local. No se puede filtrar por Bosa. Columnas disponibles:")
    print(df_local_csv.columns.tolist())
```

### Análisis de las Respuestas Seleccionadas en Bosa

```python
# Análisis y visualización de 'Primera respuesta seleccionada' en Bosa
response_col_1 = 'Primera respuesta seleccionada'
if response_col_1 in df_bosa_local.columns:
    print(f"\nDistribución de '{response_col_1}' en Bosa:")
    response_counts_1 = df_bosa_local[response_col_1].astype(str).value_counts()
    display(response_counts_1)

    # Plotting part
    plt.figure(figsize=(12, 7))
    sns.barplot(y=response_counts_1.index, x=response_counts_1.values, palette='viridis', hue=response_counts_1.index, legend=False)
    plt.title(f"Distribución de '{response_col_1}' en Bosa")
    plt.xlabel('Conteo')
    plt.ylabel('Respuesta Seleccionada')
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
else:
    print(f"La columna '{response_col_1}' no se encontró en los datos de Bosa.")
```

```python
# Análisis y visualización de 'Segunda respuesta seleccionada' en Bosa
response_col_2 = 'Segunda respuesta seleccionada'
if response_col_2 in df_bosa_local.columns:
    print(f"\nDistribución de '{response_col_2}' en Bosa:")
    # Manejar posibles valores NaN en la columna antes de value_counts
    response_counts_2 = df_bosa_local[response_col_2].fillna('NaN').astype(str).value_counts()
    display(response_counts_2)

    # Plotting part
    plt.figure(figsize=(12, 7))
    sns.barplot(y=response_counts_2.index, x=response_counts_2.values, palette='magma', hue=response_counts_2.index, legend=False)
    plt.title(f"Distribución de '{response_col_2}' en Bosa")
    plt.xlabel('Conteo')
    plt.ylabel('Respuesta Seleccionada')
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
else:
    print(f"La columna '{response_col_2}' no se encontró en los datos de Bosa.")
```

```python
# Análisis y visualización de 'Tercera respuesta seleccionada' en Bosa
response_col_3 = 'Tercera respuesta seleccionada'
if response_col_3 in df_bosa_local.columns:
    print(f"\nDistribución de '{response_col_3}' en Bosa:")
    # Manejar posibles valores NaN en la columna antes de value_counts
    response_counts_3 = df_bosa_local[response_col_3].fillna('NaN').astype(str).value_counts()
    display(response_counts_3)

    # Plotting part
    plt.figure(figsize=(12, 7))
    sns.barplot(y=response_counts_3.index, x=response_counts_3.values, palette='cividis', hue=response_counts_3.index, legend=False)
    plt.title(f"Distribución de '{response_col_3}' en Bosa")
    plt.xlabel('Conteo')
    plt.ylabel('Respuesta Seleccionada')
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
else:
    print(f"La columna '{response_col_3}' no se encontró en los datos de Bosa.")
```
